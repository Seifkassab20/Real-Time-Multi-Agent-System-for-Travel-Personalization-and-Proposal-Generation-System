```python
from backend.core.extraction_agent.extraction_agent import ExtractionAgent
from backend.core.extraction_agent.models import TranscriptSegment, Agent_output
from datetime import datetime
import time

from collections import defaultdict

agent = ExtractionAgent()
results = []
aggregated_entities = defaultdict(list)
total_start = time.time()


# Process each sentence as a segment
for i, segment in enumerate(result):
    segment = TranscriptSegment(
        segment_id=f"seg_{i:03d}",
        timestamp=datetime.now(),
        speaker="customer",
        text=segment
    )
    extraction = await agent.invoke(segment)
    results.append(Agent_output(**extraction))
print(results)
```

```python
from backend.core.recommendation_engine.recommendation_orchestrator import build_user_profile_from_extraction , merge_value, MERGE_RULES , recommend

final_profile = {}
recommendations = []

# Process each segment extraction individually (real-time)
for i, segment in enumerate(results):
    print(f"\n{'='*60}")
    print(f"Processing Segment {i+1}/{len(results)}")
    print(f"{'='*60}")
    
    # Update profile with this segment's information
    for field, rule in MERGE_RULES.items():
        final_profile[field] = merge_value(
            final_profile.get(field),
            getattr(segment, field),
            rule
        )
    
    # Build user profile with accumulated information
    user_profile = build_user_profile_from_extraction(final_profile)
    print(f"\nUpdated User Profile:\n{user_profile}")
    
    # Get recommendation based on current accumulated profile
    recommendation_result = recommend(user_profile)
    recommendations.append(recommendation_result)
    print(f"\nRecommendation (with info up to segment {i+1}):")
    print(recommendation_result)

print(f"\n{'='*60}")
print("All segments processed!")
print(f"{'='*60}")
```
